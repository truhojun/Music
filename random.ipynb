```python
from music21 import *
import random

notes = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
durations =[0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4]

def get_random_stream(length = 30):
    s = stream.Stream()
    for i in range(length):
        pitch = random.choice(notes)
        duration = random.choice(durations)
        s.append(note.Note(pitch,quarterLength=duration))
    return s
```

```python
s = get_random_stream()
play(s)
```
